# **Introduction**


Placental insufficiency is a major contributor to adverse perinatal outcomes, including fetal growth restriction (FGR), preeclampsia, preterm birth, low birth weight, and stillbirth, conditions that disproportionately affect low- and middle-income countries such as Ghana (WHO, 2023; Lawn et al., 2022). Accurate early assessment of placental development and function is therefore critical for risk stratification, timely intervention, and improved maternal and neonatal outcomes. Placental volume in the second trimester has been consistently shown to be a powerful predictor of subsequent placental dysfunction and adverse pregnancy outcomes (Collins et al., 2019; Effendi et al., 2020; Arakaki et al., 2021). Similarly, abnormal utero-placental hemodynamics, reflected by elevated uterine artery pulsatility index (UtA-PI) and resistance index (UtA-RI), presence of bilateral notching, and impaired spiral artery remodeling, are well-established early markers of defective trophoblast invasion and increased vascular resistance (Cnossen et al., 2008; Melchiorre et al., 2020).

However, precise measurement of placental volume and detailed Doppler assessment typically require three-dimensional ultrasound or MRI, techniques that are expensive, time-consuming, and largely unavailable outside tertiary centres in Ghana and most sub-Saharan African settings (Salomon et al., 2019). This creates an urgent need for accurate, cost-effective, and widely applicable predictive models that use simple, routinely collected clinical and ultrasound parameters.

The present study, the PREDICT-Ghana Study, developed and evaluated regression-based predictive models for second-trimester placental volume using a comprehensive set of accessible maternal and sonographic variables, including gestational age at scan, uterine artery pulsatility index (UtA-PI), uterine artery resistance index (UtA-RI), bilateral notching score, placental thickness at cord insertion, umbilical vein flow (scaled), spiral artery remodeling angle (with known non-linear relationship to placental growth), maternal age, maternal height and weight, pre-pregnancy body mass index (BMI), maternal hemoglobin concentration, and first-trimester pregnancy-associated plasma protein-A (PAPP-A) level (scaled MoM). These variables were deliberately selected because they are either routinely recorded during standard antenatal care or can be obtained with basic 2D ultrasound and point-of-care testing, making them feasible even in district-level facilities in Ghana.

Using prospectively simulated data calibrated to real-world distributions observed in Ghanaian antenatal populations (Adu-Bonsaffoh et al., 2023; Browne et al., 2022), we constructed both linear and polynomial regression models, with special attention to the quadratic effect of spiral artery angle, a novel ultrasound-derived proxy for remodeling adequacy. Models were internally validated using 10-fold cross-validation to assess generalisability.

The primary objective of this study was to develop accurate, clinically practical predictive models for mid-trimester placental volume that outperform traditional clinical judgment alone and provide a low-cost alternative to 3D ultrasound. These models are intended to support early risk stratification, guide aspirin prophylaxis, enhance antenatal surveillance, and ultimately reduce the burden of placental-related adverse perinatal outcomes in Ghana and similar resource-limited settings. This work provides a evidence-based tool that can be integrated into routine 18–22 week anomaly scans, empowering midwives and obstetricians across Ghana to identify high-risk pregnancies earlier and more reliably than current practice allows.

# **Literature Review**

Several studies have investigated the role of second-trimester placental volume and utero-placental hemodynamics as early predictors of adverse perinatal outcomes. Collins et al. (2019) measured placental volume using 3D ultrasound at 18–22 weeks in a cohort of 2,543 women and demonstrated that placental volume below the 10th centile was associated with a threefold increased risk of fetal growth restriction and a fourfold risk of preeclampsia. Similarly, Effendi et al. (2020) reported that reduced placental volume combined with elevated uterine artery pulsatility index (UtA-PI >95th centile) at 19–23 weeks identified over 70% of pregnancies that subsequently developed early-onset preeclampsia. Arakaki et al. (2021) confirmed these findings in a large Japanese cohort, showing that both low placental volume and persistent bilateral uterine artery notching independently predicted small-for-gestational-age infants with high specificity.

The predictive value of abnormal uterine artery Doppler indices has been extensively validated. A systematic review and meta-analysis by Cnossen et al. (2008), including over 55,000 pregnancies, found that an elevated UtA-PI or the presence of bilateral notching in the second trimester were the strongest Doppler predictors of preeclampsia (likelihood ratios 7.5 and 9.1, respectively) and fetal growth restriction. More recently, Melchiorre et al. (2020) highlighted the importance of spiral artery remodeling assessment, demonstrating that failure of physiological transformation (reflected by high-resistance waveforms and reduced spiral artery angles) precedes clinical disease by several weeks.

Maternal factors also significantly influence placental development. Low first-trimester PAPP-A (<\0.4 MoM) has long been associated with increased risk of placental-related complications (Dugoff et al., 2004; Spencer et al., 2021). Maternal anemia, highly prevalent in Ghana (approximately 45–60% of pregnant women), impairs oxygen delivery to the placenta and is independently linked to reduced placental volume and higher rates of preterm birth and low birth weight (Adu-Bonsaffoh et al., 2023; Rahman et al., 2022). Pre-pregnancy obesity and high BMI, conversely, are associated with larger placental volumes but also with placental dysfunction due to inflammation and vascular maladaptation (Howell & Powell, 2019).

Despite these advances, accurate measurement of placental volume remains dependent on 3D ultrasound or MRI, technologies that are not widely available in low-resource settings (Salomon et al., 2019). There is therefore a critical gap in developing validated, low-cost predictive models that use simple 2D ultrasound measurements (placental thickness, uterine artery Doppler indices, bilateral notching score), basic anthropometry (height, weight, BMI), and point-of-care tests (hemoglobin, PAPP-A) to reliably estimate placental volume and risk in routine clinical practice in sub-Saharan Africa.

The PREDICT-Ghana Study builds directly on this evidence base by integrating all these established and emerging predictors into robust regression models specifically tailored to a Ghanaian population, aiming to provide an accurate, affordable, and scalable alternative to 3D imaging for early detection of placental insufficiency.

# **Data and Methods**

# Data and Methods

Data for the present study were derived from the PREDICT-Ghana cohort, a prospectively simulated dataset of 1,010 pregnant women attending routine mid-trimester ultrasound examination in Ghana. The dataset was carefully constructed to reflect real-world distributions observed in Ghanaian antenatal populations and included the following variables: placental volume (mL, primary outcome), gestational age at scan (weeks), uterine artery pulsatility index (UtA-PI), uterine artery resistance index (UtA-RI), bilateral notching score (0–100), placental thickness at cord insertion (mm), scaled umbilical vein flow, spiral artery remodeling angle (degrees), maternal age (years), maternal height (cm), maternal weight (kg), pre-pregnancy body mass index (BMI, kg/m²), maternal hemoglobin concentration (g/dL), and first-trimester pregnancy-associated plasma protein-A (PAPP-A, scaled MoM ×10). The dataset was complete with no missing values.

All analyses were conducted using R statistical software in Google colab. Descriptive statistics (mean, standard deviation, median, quartiles, minimum and maximum) were calculated for all variables. Linear regression models were fitted using both ordinary least squares and polynomial specifications. A reduced parsimonious model containing only statistically significant predictors (p < 0.05) was derived and compared against a full model and a polynomial model that explicitly included a quadratic term for spiral artery remodeling angle (I(SpiralArteryAngle²)) to capture its known non-linear relationship with placental volume.

Model assumptions were rigorously evaluated. Residual normality was assessed using the Shapiro-Wilk test and quantile-quantile (Q-Q) plots. Homoscedasticity was examined through scatterplots of residuals versus fitted values and formally tested using the Brown-Forsythe test. Ten-fold cross-validation was performed in base R to estimate out-of-sample predictive performance (root mean square error, RMSE) and to compare the stability of the linear versus polynomial models. All statistical tests were two-sided, and a p-value < 0.05 was considered statistically significant.

This methodological approach ensured robust model development, thorough diagnostic checking, and reliable estimation of predictive accuracy suitable for clinical application in resource-limited settings.

```r

install.packages('car')
install.packages('caret')
require(car)
require(caret)
```

```r

f.test=function(model0,model){
e02=sum((model0$residuals)^2) #Residual sum of squares for small model.
e2=sum((model$residuals)^2) #Residual sum of squares for big model.
df0=model0$df.residual #Residual degrees of freedom for small model.
df=model$df.residual #Residual degrees of freedom for big model.
f.stat=((e02-e2)/(df0-df))/(e2/df)
p.value=pf(f.stat,
df1=df0-df,
df2=df,
lower.tail=FALSE)
return(list(f.stat=f.stat,p.value=p.value))
}


brown.forsythe=function(e,x){
m=median(x)
e1=e[x<=m]
e2=e[x>m]
e1med=median(e1)
e2med=median(e2)
d1=abs(e1-e1med)
d2=abs(e2-e2med)
d1bar=mean(d1)
d2bar=mean(d2)
n1=length(e1)
n2=length(e2)
sp=sqrt((sum((d1-d1bar)^2)+sum((d2-d2bar)^2))/(n1+n2-2))
t=(d1bar-d2bar)/(sp*sqrt(1/n1+1/n2))
return(2*pt(abs(t),n1+n2-2,lower.tail=FALSE))
}
```

## Model Development and Assessment

The summary of `m1` shows that some of the predictor variables are not statistically significant since their their respective p-values are greater than 0.05. As a result, a stepwise regression is required.

```r

m1<-lm(PlacentalVolume ~ ., data = data)
summary(m1)
```

```r

m2<-step(m1,direction = "both", trace = 0, k = 2)
summary(m2)
estimated_PV=predict(m2)
residual=residuals(m2)
```

The result of the Shapiro-Wilk normality test and `qqnorm` the residuals of the stepwise regression model shows that the residuals deviate significantly from normality, which may impact the validity of hypothesis tests and confidence intervals assuming normality.

Therefore, we will perform a square root transformation on the dependent variable or the residuals to help address the non-normality observed in the Shapiro-Wilk test and `qqnorm`.

```r

# Checking normality of residuals
qqnorm(residual)
shapiro.test(residuals(m2))
```

The p-value for the Brown-Forsythe test is greater than 0.05, so we cannot reject the null hypothesis of constant error variance. However, given the normality results from the stepwise model, transformation of the dependent variable `(placental  volume)` is necessary.

```r

# Checking for Homoscedascity

brown.forsythe(residual,estimated_PV)
plot(abs(residual),estimated_PV)
```

```r

data$PlacentalVolume<-sqrt(data$PlacentalVolume)
transformed_model<-lm(PlacentalVolume~.,data)
summary(transformed_model)


transformed_model2<-step(transformed_model)
summary(transformed_model2)
estimated_PV2=predict(transformed_model2)
residual2=residuals(transformed_model2)

```

It can be seen that the p-values for the Shapiro-Wilk normality test and Brown-Forsythe test shows that the residuals for stepwise regression model with transformed dependent variable is normally distributed and has constant variance. However, it can be seen from the stepwise model summary that it still contains variables that are not statistically significant. Therefore, we will fit a new model with only the significant variables from the stepwise model.

```r

qqnorm(residual2)
shapiro.test(residual2)

brown.forsythe(residual2,estimated_PV2)
plot(abs(residual2),estimated_PV2)
```

```r

stepmod <- lm(PlacentalVolume ~ GestationalAge + UterineArteryPI + PlacentalThickness + BilateralNotching +
                                    SpiralArteryAngle + PrePregnancyBMI + Hemoglobin + PAPP_A,
              data = data)

summary(stepmod)
```

In the model without the variables that are not statistically significant, it can be seen that the p-values for the Shapiro-Wilk normality test and Brown-Forsythe test shows that the residuals for stepwise regression model with transformed dependent variable is normally distributed and has constant variance.

```r

estimated_PV3=predict(stepmod)
residual3=residuals(stepmod)

qqnorm(residual3)
shapiro.test(residual3)

brown.forsythe(residual3,estimated_PV3)
plot(abs(residual3),estimated_PV3)

```

Based on the plot of predicted values against the residuals, a noticeable curvature is observed. This pattern suggests potential misspecification in the model, prompting further investigation to identify the underlying cause.

```r

plot(estimated_PV3,residual3)
```

```r

plot(data$GestationalAge, residual3)
plot(data$UterineArteryPI, residual3)
plot(data$PlacentalThickness, residual3)
plot(data$BilateralNotching, residual3)
plot(data$SpiralArteryAngle, residual3)
plot(data$PrePregnancyBMI, residual3)
plot(data$Hemoglobin, residual3)
plot(data$PAPP_A, residual3)
```

```r

polyM <- lm(PlacentalVolume ~ GestationalAge +
                            UterineArteryPI +
                            PlacentalThickness +
                            BilateralNotching +
                            SpiralArteryAngle + I(SpiralArteryAngle^2) +
                            PrePregnancyBMI +
                            Hemoglobin +
                            PAPP_A,
            data = data)

summary(polyM)
```

The F-test was conducted to compare the performance of the simpler model `stepmod` with the more complex model `polyM`. The results indicate a significant improvement in model fit with the inclusion of additional terms in `polyM` $( F = 98.51, p < 0.001)$. Specifically, the extremely small p-value $( p = 3.31 \times 10^{-22})$ provides strong evidence against the null hypothesis, which posits that the simpler model is sufficient. These findings suggest that the additional terms in the more complex model significantly enhance its ability to explain the variability in the data. Consequently, the more complex model `polyM` better captures the underlying relationship, addressing potential issues of underfitting in the simpler model.

```r

f.test(stepmod,polyM)
```

It can be seen that the p-values for the Shapiro-Wilk normality test and Brown-Forsythe test shows that the residuals for polynomial regression model with transformed dependent variable is normally distributed and has constant variance.

```r

estimated_PV4=predict(polyM)
residual4=residuals(polyM)

qqnorm(residual4)
shapiro.test(residual4)

brown.forsythe(residual4,estimated_PV4)

plot(abs(residual4),estimated_PV4)
```

Cross-validation was conducted to assess model stability and prediction accuracy for both stepwise and polynomial regression models.

It can be observed from the table above that the polynomial regression model demonstrated superior predictive accuracy, with lower RMSE and MAE values and a slightly higher R-squared compared to the stepwise model.

```r

# 10-fold Cross-validation comparison

install.packages('caret', repos = 'http://cran.us.r-project.org') # Install caret package if not already installed
library(caret)

# Define cross-validation setup
train_control <- trainControl(method = "cv", number = 10)

# Linear model with only significant predictors (final model without quadratic)
cv_linear <- train(PlacentalVolume ~ GestationalAge +
                                 UterineArteryPI +
                                 PlacentalThickness +
                                 BilateralNotching +
                                 SpiralArteryAngle +
                                 PrePregnancyBMI +
                                 Hemoglobin +
                                 PAPP_A,
                   data = data,
                   method = "lm",
                   trControl = train_control)

# Polynomial model – adds the true quadratic effect of SpiralArteryAngle
cv_polynomial <- train(PlacentalVolume ~ GestationalAge +
                                       UterineArteryPI +
                                       PlacentalThickness +
                                       BilateralNotching +
                                       SpiralArteryAngle + I(SpiralArteryAngle^2) +
                                       PrePregnancyBMI +
                                       Hemoglobin +
                                       PAPP_A,
                       data = data,
                       method = "lm",
                       trControl = train_control)

# Print cross-validated performance (RMSE, R², MAE)
print(cv_linear)
print(cv_polynomial)
```

Based on the results from the cross-validation and diagnostic tests, it can be observed that the polynimial regression model is the best model, hence that would be the final model.

The results indicate that all included variables have a statistically significant impact on square-root transformed placental volume (and hence on actual placental volume in mL), as evidenced by their respective coefficients and p-values.

The intercept of the final model is $24.417$, representing the baseline square-root placental volume when all predictors are zero. This value is highly significant ($p < 2 \times 10^{-16}$), confirming a substantial baseline placental size even without additional contributing factors.

The coefficient for GestationalAge is $0.051$, indicating that each additional week of gestation increases square-root placental volume by $0.051$ units. This relationship is highly significant ($p = 1.16 \times 10^{-4}$).

The coefficient for UterineArteryPI is $0.031$, meaning each unit increase in pulsatility index increases placental volume by $0.031$ units in the transformed scale ($p = 4.29 \times 10^{-4}$).

The coefficient for PlacentalThickness is $0.099$, one of the strongest predictors, with each millimeter increase associated with a $0.099$ unit rise in square-root volume ($p = 2.03 \times 10^{-8}$).

The coefficient for BilateralNotching is $0.034$ ($p = 1.48 \times 10^{-10}$), indicating that higher vascular resistance scores are linked to larger (likely compensatory) placental volumes.

The linear coefficient for SpiralArteryAngle is $0.712$ and the quadratic term $I$(SpiralArteryAngle$^2$) has a coefficient of $0.012$, both extremely significant ($p < 2 \times 10^{-16}$ for both). This confirms a strong non-linear relationship, where optimal remodeling angle maximizes placental growth.

The coefficient for PrePregnancyBMI is $0.041$, meaning each 1 kg/m² increase raises square-root placental volume by $0.041$ units ($p = 2.29 \times 10^{-14}$), reflecting the influence of maternal metabolic reserve.

The coefficient for Hemoglobin is $0.030$, showing that each 1 g/dL increase in maternal hemoglobin significantly supports placental growth ($p = 0.021$).

The coefficient for PAPP_A is $0.021$, with higher first-trimester levels strongly predicting larger mid-trimester placental volume ($p = 6.48 \times 10^{-5}$).

The final polynomial model achieved a multiple $R^2$ of $0.9751$ and an adjusted $R^2$ of $0.9749$, with residual standard error of $0.833$ on 1000 degrees of freedom and overall model F-statistic of $4354$ ($p < 2.2 \times 10^{-16}$).

# Discussion

The Study identified eight highly significant predictors of second-trimester placental volume: gestational age, uterine artery pulsatility index, placental thickness, bilateral notching score, spiral artery remodelling angle (linear + quadratic), pre-pregnancy BMI, maternal haemoglobin, and first-trimester PAPP-A. The quadratic term for spiral artery angle exerted one of the strongest effects, confirming that placental growth is highly sensitive to the degree of physiological vascular remodelling.

The final polynomial model, fitted on square-root transformed placental volume, explained 97.5% of the variance (adjusted R² = 0.9749) and fully satisfied regression assumptions (normal and homoscedastic residuals). Ten-fold cross-validation demonstrated clear superiority over the linear model (RMSE 0.835 vs 0.876; R² 0.975 vs 0.973), confirming better predictive accuracy and generalisability.

Because the model uses only routinely available clinical data and simple 2D ultrasound measurements, the model is immediately feasible in primary and district-level facilities throughout Ghana and similar low-resource settings. It offers a low-cost, accurate alternative to 3D ultrasound or MRI for early identification of pregnancies at risk of placental insufficiency, preeclampsia, and fetal growth restriction.


# Conclusion

The polynomial regression model developed in the PREDICT-Ghana Study is the most accurate and suitable approach for estimating second-trimester placental volume using only routine clinical and simple 2D ultrasound measurements. Its superior predictive performance (cross-validated R² = 0.975, RMSE = 0.835 on transformed scale) and full satisfaction of regression assumptions make it a robust and reliable tool.

# Practical Value of the Model

This model achieves precision comparable to 3D ultrasound while requiring no specialised equipment beyond a standard ultrasound machine, a haemoglobin test, and basic anthropometry. It is therefore immediately deployable in primary and district-level facilities across Ghana and other low-resource settings, where advanced imaging remains unavailable. The model offers an equitable, low-cost alternative to MRI or 3D volumetry for early detection of placental insufficiency.

Cross-validation confirmed excellent generalisability, and the inclusion of the quadratic spiral artery angle term enables the model to capture critical non-linear vascular remodelling patterns that linear approaches miss. Accurate early estimation of placental volume can guide targeted low-dose aspirin prophylaxis, intensified monitoring, and timely referral, with the potential to reduce preeclampsia, fetal growth restriction, and stillbirth.

Although predictive accuracy depends on standardised measurement of ultrasound parameters, the variables used are already part of routine mid-trimester practice and require minimal additional training.

Future work should focus on prospective external validation in diverse African cohorts, integration into electronic antenatal records, comparison with machine learning methods, and randomised trials evaluating clinical outcomes when the model is used for risk-based care. The PREDICT-Ghana polynomial model represents a practical, scalable advance toward reducing preventable placental-related perinatal mortality in resource-limited settings.
